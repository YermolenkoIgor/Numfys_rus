### Вступление

В этом примере мы обсуждаем физику классического парамагнетика и модель Изинга в одном и двух измерениях и попутно затрагиваем общие аспекты статистической механики. 

Модель Изинга - это математическая модель в статистической механике, которая может описать явление ферромагнетизма. Модель состоит из дискретных переменных, представляющих магнитные дипольные моменты атомов $\textit{spins}$, которые могут находиться в одном из двух состояний (+1 или -1). Спины могут быть расположены в виде решетки, позволяя каждому вращению взаимодействовать со своими соседями. Модель позволяет идентифицировать фазовые переходы, как упрощенную модель реальности. Двумерная модель Изинга квадратной решетки является одной из простейших статистических моделей, показывающих фазовый переход [1].

Модель XY является важной спиновой моделью легких плоских анизотропных магнитных изоляторов, где спины $\mathbf{S}_i$ с двумя компонентами $\mathbf{S}_i = \left(S_{xi}, S_{yi}, 0\right)$ находятся на множестве точек решетки [2]. (Это также широко используемое эффективное описание сверхтекучих жидкостей, таких как He$^{4}$ при низких температурах, которая представляет собой систему конденсированного вещества с переносом вещества без диссипации! [3]) Здесь мы принимаем решетку за трехмерную кубическую решетку. Гамильтониан этой спиновой системы задается

\begin{align*}
H = -J \sum \limits_{\left< i,j \right>} \mathbf{S_i} \cdot \mathbf{S_j} - \mathbf{h} \cdot \sum\limits_{i=1}^{N} \mathbf{S_i},
\end{align*}

где $\left< i, j \right>$ означает, что узлы решетки $i$ и $j$ являются ближайшими соседями, а $J$ - это энергия, которая связывает вместе вращения на узлах решетки ближайших соседей. Вектор $\textbf{h}$ представляет внешнее магнитное поле.

### Аналитическое решение парамагнетика XY

Рассмотрим набор $N$ независимых XY спинов $\mathbf{S}_i = \left(S_{xi}, S_{yi}, 0\right)$ во внешнем магнитном поле $\mathbf{h} = \left(h, 0, 0\right)$, таких что $\mathbf{h} \cdot \mathbf{S}_i = hS\cos(\phi_i)$, где $\phi_i$ - угол между спином и внешним магнитным полем, а $S$ это длина спина, которую мы считаем фиксированной. Мы устанавливаем $J$ = 0, так что все вращения независимы. Гамильтониан этой системы равен

\begin{align*}
H = - \mathbf{h} \cdot \sum\limits_{i=1}^{N} \mathbf{S_i} = -hS \sum\limits_{i=1}^{N} \cos(\phi_i),
\end{align*}

#### Каноническая статистическая сумма $Z$

Для определения термодинамических величин, таких как давление, плотность энергии и энтропия системы, мы вычисляем *статфункцию* $Z$, в нашем случае классическую каноническую, заданную в общем виде как

\begin{align*}
Z = \sum\limits_{i=1} e^{-\beta E_i},
\end{align*}

где $\beta$ - это так называемый *множитель Лагранжа*, и можно найти, что он равен $1/k_B T$. (Это вытекает из нашей цели найти наиболее вероятное распределение энергии с учетом определенных ограничений. Более подробную информацию можно найти в любом стандартном учебнике по статистической механике, например [4].) Статсумма $Z$ является одной из наиболее важных величин в равновесной статистической механике. Это сумма по всем микросостояниям $i$. В квантовой механике индекс $i$ обычно является дискретной переменной, а функция разбиения-безразмерной переменной. В классической механике переменная $i$ часто становится непрерывной, а сумма заменяется интегралом. В нашем случае $\phi \in [0, 2\pi)$ является непрерывной переменной, поэтому, вставляя наше выражение для $H$ в выражение для $Z$, мы получаем

\begin{align*}
Z &= \sum\limits_{\{\mathbf{S_i}\}} e^{-\beta H} = (Z_1)^N \\
Z_1 &= \sum\limits_{\mathbf{S}} e^{\beta \mathbf{h} \cdot \mathbf{S}} = \int_0^{2\pi} e^{\beta h S \cos\phi} d\phi= 2\pi I_0(\beta h),
\end{align*}

где $I_0$ - модифицированная функция Бесселя первого рода и нулевого порядка. На последнем шаге мы установили $S = 1$ для удобства. Более подробную информацию об измененной функции Бесселя см. в приложении. Мы строим модифицированную функцию Бесселя порядков от 0 до 3, чтобы помочь нашей интуиции. Bмпортируем необходимые пакеты и устанавливаем общие параметры рисунка. Мы используем реализацию модифицированной функции Бесселя, предоставленной в библиотеке `scipy`, `scipy.special.i0` для порядка 0 и `scipy.special.iv` для более высоких порядков $v$.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.special as sci
import time
import itertools as it
from matplotlib import rc
plt.style.use('bmh')
```

```python
# --- Set common figure parameters
newparams = {'font.size': 14}
plt.rcParams.update(newparams)
```

```python
plt.figure(figsize=(16, 4))
x = np.linspace(0, 4, 100)
plt.plot(x, sci.i0(x));
plt.plot(x, sci.iv(1, x));
plt.plot(x, sci.iv(2, x));
plt.plot(x, sci.iv(3, x));
plt.title('Modified Bessel function')
plt.xlabel('$x$')
plt.ylabel('$I_v$')
plt.legend(['$I_0$', '$I_1$', '$I_2$', '$I_3$'], loc=2)
plt.show()
```

#### Средняя намагниченность на спин $m(h, T)$

Теперь, когда мы знаем $Z$, мы решаем вычислить компонент $M$ средней намагниченности, параллельной внешнему полю, и удельную теплоемкость $C_h$ (при постоянном магнитном поле) нашей системы. Пусть $m(h, T)$ обозначает среднюю намагниченность на спин, которая определяется


\begin{align*}
m &\equiv \frac{1}{N} \left<\hat{h} \cdot \mathbf{M}\right> \\
&= \frac{1}{N} \left<\hat{h} \cdot \sum_i \mathbf{S}_i\right> \\
\left<\hat{h} \cdot \mathbf{M}\right> 
&= \frac{\partial(\ln{Z})}{\partial(\beta h)} = N \frac{I_0'(\beta h)}{I_0(\beta h)} = Nm
\\
\Rightarrow m(h, T) &= \frac{I_1(\beta h)}{I_0(\beta h)},
\end{align*}

где мы использовали отношение $I_0' = I_1$.
Особый интерес представляет исследование значений $m$, когда $\beta h S \ll 1$.
Сюжет полезен, когда мы хотим получить немного физического понимания:

```python
plt.figure(figsize=(16, 4))
beta_hs = np.linspace(0, 16, 200)
plt.plot(beta_hs, sci.i1(beta_hs)/sci.i0(beta_hs));
plt.title('Magnetisation per spin in the XY model')
plt.xlabel('$\\beta h$')
plt.ylabel('$m$')
plt.show()
```

Мы наблюдаем, что $m \to 0 $ при $\beta h \to 0$. Это тот случай, когда $h \to 0$ или $T \to \infty$. Первый предел означает, что средняя намагниченность исчезает, когда мы удаляем внешнее поле; все микросостояния имеют одинаковую энергию и, следовательно, равные вероятности, что приводит к тому, что средняя намагниченность стремится к нулю. Последний предел означает, что по мере повышения температуры тепловые флуктуации начнут доминировать над выравнивающим эффектом внешнего поля.

Функция $m(h, T)$ аналитична везде, включая $(0,0)$, следовательно, $\lim\!_{(h, T)\to(0,0)}\; m(h,T) = 0$ для любого пути мы позволяем $(h,T)$ приблизиться к началу координат. Другими словами, невозможно сохранить чистую намагниченность после удаления внешнего поля - точно так же, как можно было бы ожидать, когда нет связи между спинами!

#### Удельная теплоемкость на спин $C_h$

Чтобы вычислить $C_h$, мы сначала находим энтальпию $H_E = \langle H \rangle$ нашей системы

\begin{align*}
H_E &= - \frac{\partial (\ln{Z})}{\partial \beta} = -h N \frac{I_1(\beta h)}{I_0(\beta h)} \\
C_h &= \frac{1}{N} \left(\frac{\partial H_E}{\partial T}\right)_h = -h \frac{\partial(I_1/I_0)}{\partial (\beta h)} \frac{\partial (\beta h)}{\partial T} \\
&= \left(\frac{-h^2}{-k_B T^2}\right) \cdot \left(\frac{I_1'}{I_0} - \frac{I_1^2}{I_0^2}\right) \\
&= k_B h^2 \frac{\beta^2}{I_0^2} (I_1'I_0 - I_1^2).
\end{align*}

Используя рекурсивные соотношения, мы находим

\begin{align*}
I_1' &= \frac{d}{dx}(xI_1) = I_1 + xI_1' = xI_0, \\
I_1' &= \frac{1}{x}(xI_0 - I_1), \\
I_1'I_0 - I_1^2 &= I_0(I_0 - \frac{I_1}{x}) - I_1^2, \\
x \rightarrow 0 : I_0 &\rightarrow 1, \\
x \ll 1 : I_0 &\simeq 1, \\
I_1 &\simeq \frac{x}{2}, \\
\implies C_h &\simeq \frac{1}{2} k_B (\beta h)^2.
\end{align*}

Что происходит, когда $h \to 0$? Мы получаем, что $C_h \to 0$! Причина этого заключается в следующем: Другой способ записи удельной теплоты заключается в терминах колебаний энергии

\begin{align*}
C_h &= k_B \beta^2 \left[\langle H^2 \rangle - \langle H \rangle^2\right], \\
H &= - \mathbf{h} \cdot \sum\limits_{i=1}^N \mathbf{S}_i.
\end{align*}

$h \to 0 \implies H \to 0 \implies \text{отсутствие колебаний энергии} \implies \text{отсутствие удельной теплоты }$ $C_h$.

### Численное решение модели Изинга в одном и двух измерениях

Теперь мы перейдем от обсуждения аналитического решения парамагнетика к численному решению модели Изинга в одном и двух измерениях. Это простая, но очень важная модель взаимодействующих спинов $N$, которые находятся в точках некоторой решетки. Таким образом, он может служить моделью магнитных изоляторов, но также имеет широкий спектр применений во многих других системах. Поскольку многие из его свойств точно известны в одном и двух измерениях (и с большой точностью также в трех измерениях, основанных на численной работе), он служит полезным ориентиром для схем аппроксимации и вычислительных схем. В этой задаче мы будем численно изучать модель на одномерной цепочке
и на двумерной квадратной решетке с периодическими граничными условиями. Модель определяется гамильтонианом

\begin{align*}
H = -J \sum\limits_{\left< i,j \right>} s_i s_j - h \sum\limits_i s_i,
\end{align*}

где $\left< i,j \right>$ означает, что участки решетки $i$ и $j$ являются ближайшими соседями, $J > 0$ - это энергия, которая связывает вместе вращения на ближайших соседних участках решетки, а $h$ - внешнее магнитное поле. Без потери общности мы устанавливаем $h > 0$. Наконец, $s_i$ - это спиновые переменные системы, которые могут принимать значения $s_i =\pm 1$.

Каноническая статфункция этой системы затем задается

\begin{align*}
Z = \sum\limits_{\{s_i\}} e^{-\beta H}.
\end{align*}

Эта модель может быть изучена с двумя различными условиями на граничных спинах:

i) Открытые граничные условия. Здесь спины на краю решетки взаимодействуют только со спинами внутри решетки.

ii) Периодические граничные условия. Здесь спины на краю решетки взаимодействуют со спинами внутри решетки и со спинами в той же строке или столбце на противоположном краю. (Это соответствует свертыванию двумерной квадратной решетки в тор).

В этом примере мы используем периодические граничные условия.

#### Вычисление величин $Z$, $U$, $C_h$ и $m$ (на спин)

Мы хотим вычислить статсумму $Z$, внутреннюю энергию $U$, удельную теплоемкость $C_h$ и среднюю намагниченность $m$ для одномерной решетки и квадратных решеток с размерами $2 \times 2$, $3 \times 3$, $4 \times 4$ как функции температуры. Вычисления выполняются с параметрами $N=16$ и $h/J \in \{0, 1/2, 1\}$, а затем строятся термодинамические величины.

```python
def generate_nn_array(N, D):
    """Generate array of nearest neighbour (nn) indices.
    D=1: This is pretty straightforward, spin i interacts with spin i+1 mod N
    D=2: This is a bit trickier. Even though we have two dimensions, we
    represent our grid as a one dimensional array (a vector). The grid may be
    written as (for a 3x3 lattice, the number is the index of the array:

            6 7 8
            3 4 5
            0 1 2

    This means that we need an N x 2 array which contains the indices of the
    nn in the positive x and y directions. For instance, nn[1, 0] should be 
    equal to 1, nn[0, 1] should be equal to 3.

    :N: number of lattice sites in total (not linear size)
    :D: lattice dimension (if D=2, make sure N is a perfect square)
    :returns: nn array (number of sites, direction 0:+ 1:-)
    """
    nns = np.zeros((N, D), dtype=int)
    if D == 1:
        nns[:, 0] = [(i+1)%N for i in range(N)]
    elif D == 2:
        stride = int(np.sqrt(N))
        for i in range(N):
            nns[i, 0] = (i+1)%stride + np.floor(i/stride)*stride
            nns[i, 1] = (i+stride)%N
    else:
        print('D not supported')
        return None
    return nns


def hamiltonian(conf, J, h, nns, N, D):
    """Calculate the Hamiltonian for a given configuration by looping over each
    lattice site and calculating s_i*s_j for each nearest neighbor, as well as
    the external field contribution h.

    :conf: array of a spin configuration
    :returns: value of H
    """
    H = 0.
    for i in range(N):
        for j in range(D):
            # --- nn term in direction j
            H += -J * conf[i] * conf[nns[i, j]]
        # --- external field contribution
        H += -h*conf[i]
    return H


def compute_quantities(N=4, D=2, J=1., h=0.5, nt=201, ts=0.5, te=5.):
    """Compute the thermodynamic quantities.
    :N: number of lattice sites
    :D: number of dimensions (1 & 2 supported)
    :J: coupling strength
    :h: external magnetic field strength
    :nt: number of temperature steps
    :ts: start temperature
    :te: end temperature
    :returns: a tuple containing the thermodynamic quantities
              (temps, lnZ, Ch, U, mag).
    """
    nns = generate_nn_array(N, D)

    # --- Initialize arrays for measurements
    temps = np.linspace(ts, te, num=nt, endpoint=True)
    lnZ = np.zeros(nt)
    mag = np.zeros(nt)
    U = np.zeros(nt)
    Ch = np.zeros(nt)
    
    # --- Start clock for timing
    st = time.time()

    # --- Create all possible configurations of +- spins on N sites.
    # --- it.product is a function from the itertools package that creates an iterable
    # --- (an object that can be looped over) which successively creates all possible
    # --- combinations of +-1
    configurations = it.product([1, -1], repeat=N)
    # --- Loop over configurations
    for c in configurations:
        # --- Calculate the value of H
        ham = hamiltonian(c, J, h, nns, N, D)
        # --- Loop over temperatures
        # for i, t in enumerate(temps):
        # --- Calculate contribution to the partition function and observables
        # --- from the given configuration and add to the appropriate arrays.
        # --- Here we use vector operations, as it is more effective.
        bman = np.exp(-ham/temps)
        lnZ += bman
        # --- <H>
        U += ham*bman
        # --- <H^2>
        Ch += ham**2*bman
        # --- <m>
        mag += sum(c)*bman

    runtime = (time.time() - st)/len(temps)
    print('Average of {0:4.3f} s per temp iteration.'.format(runtime))

    # --- Normalize results
    mag /= (lnZ*N)
    U /= lnZ
    Ch = (Ch/lnZ-U**2)/temps**2/N
    U /= N
    lnZ = np.log(lnZ)
    
    return (temps, lnZ, Ch, U, mag)


def generate_plots(N=4, D=2, nt=201, ts=0.5, te=5.):
    """Compute and plot the thermodynamic quantities for h/J = 0, 0.5 and 1.
    """ 
    fig = plt.figure(figsize=(16, 10))   
    fig.suptitle('Parameters: D = {0}, N = {1}'.format(D, N), y=1.0)
    ax1 = fig.add_subplot(221)
    ax2 = fig.add_subplot(223)
    ax3 = fig.add_subplot(222)
    ax4 = fig.add_subplot(224)
    
    J = 1.0
    h_values = [0.0, 0.5, 1.0]
    
    for h in h_values:
        (temps, lnZ, Ch, U, mag) = compute_quantities(N, D, J, h, nt, ts, te)
    
        ax1.plot(temps, lnZ)
        ax1.set_ylabel(r'$\ln Z$')
        ax1.set_xlabel(r'$k_B T$')

        ax2.plot(temps, U)
        ax2.set_ylabel(r'$U$')
        ax2.set_xlabel(r'$k_B T$')
    
        ax3.plot(temps, Ch)
        ax3.set_ylabel(r'$C_h$')
        ax3.set_xlabel(r'$k_B T$')

        ax4.plot(temps, mag)
        ax4.set_xlabel(r'$k_B T$')
        ax4.set_ylabel(r'$\langle m \rangle$')

    legend_info = ['h = {0}'.format(h) for h in h_values]
    ax1.legend(legend_info)
    ax2.legend(legend_info, loc='lower right')
    ax3.legend(legend_info)
    ax4.legend(legend_info)
    fig.tight_layout()
```

```python
start_time = time.time()

generate_plots(N=16, D=1)
generate_plots(N=16, D=2)

print("Program runtime: %.5s s." % (time.time() - start_time))
```

Наблюдательный читатель, возможно, заметил, что эти значения $k_B T$ соответствуют смехотворно высоким температурам.
Однако нет необходимости беспокоиться, так как в более реалистичной ситуации гамильтониан, как правило, масштабируется коэффициентом порядка, сопоставимым с коэффициентом боровского магнетона $\mu_B = 9,27 \cdot 10^{-24} \mathrm{J/T}$.
Мы обнаруживаем, что удельная теплоемкость имеет отчетливый пик при низких температурах, который смещается вправо по мере увеличения $h$. Этот тип пика обычно называют аномалией Шоттки [5]. Аномалия возникает из-за энергетического разрыва между основным состоянием и первым возбужденным состоянием:  

Пусть $\Delta E$ обозначает размер этой энергетической щели. При низких температурах $k_B T \ll \Delta E$ тепловая энергия недостаточно велика, чтобы преодолеть энергетический разрыв $\implies$ несколько доступных энергетических состояний $\implies$ небольшая $C_h$. Когда $k_B T \sim \Delta E$, происходит внезапное увеличение числа доступных энергетических состояний, следовательно, внезапное увеличение $C_h$. Удельная теплоемкость будет снова снижаться по мере дальнейшего повышения температуры из-за фактора $1/k_B T^2$ и конечного числа энергетических состояний, что приведет к наблюдаемому пику. Когда мы увеличиваем $h$, мы также увеличиваем энергетический разрыв, и пик смещается в сторону более высоких температур. Это объясняет тенденцию, которую мы наблюдаем на графиках.  

Удельная теплоемкость в крайних пределах температуры стремится к нулю. Существует только одно доступное энергетическое состояние в пределе $T \to 0$, а именно основное состояние, которое подразумевает, что нет флуктуаций энергии и, следовательно, $C_h = 0$. Когда $T$ приблизится к бесконечности, спины постепенно станут более хаотично ориентированными, и поэтому намагниченность и внутренняя энергия будут приближаться к нулю. Скорости, с которыми $U$ и $m$ изменяются по отношению к $T$, также падают при более высоких температурах, потому что случайность спинов становится менее чувствительной к изменениям в $T$, когда температура становится очень высокой.  
Другое выражение для удельной теплоты - $C_h = \frac{\partial \langle E \rangle}{\partial T}$, и мы, таким образом, видим, что $C_h \to 0$ как $T \to \infty$. На первый взгляд может показаться нелогичным, что внутренняя энергия стремится к нулю как $T \to \infty$, но помните, что мы не учитываем никаких форм кинетической энергии.

#### Удельная теплоемкость $C_h$ в 1D и 2D модели

Кроме того, мы позволим $h = 0$ и исследуем вычисленные значения $C_h$ как в одномерном, так и в двумерном случае для $N \in \{4, 9, 16\}$.

```python
start_time = time.time()

fig = plt.figure(figsize=(10, 5))
N_values = [4, 9, 16]

for N in N_values:
    (temps, lnZ, Ch, U, mag) = compute_quantities(N, D=1, J=1.0, h=0, ts=0.1, te=6.0)
    plt.plot(temps, Ch, '--')
    (temps, lnZ, Ch, U, mag) = compute_quantities(N, D=2, J=1.0, h=0, ts=0.1, te=6.0)
    plt.plot(temps, Ch)
    
plt.xlabel('$k_B T$')
plt.ylabel('$C_h$')
plt.title('Specific heat')
legend_info = ['N = {0}, D = {1}'.format(N, D) for N in N_values for D in [1, 2]]
plt.legend(legend_info)
print("Program runtime: %.5s s." % (time.time() - start_time))
```

Пики в 2D-модели происходят при более высоких температурах, чем в 1D-модели. Это можно объяснить следующим образом: в 2D-модели больше спинов ближайших соседей, что означает, что для переключения спина требуется больше энергии, увеличивая энергетический зазор между основным состоянием и первым возбужденным состоянием.

#### Модифицированная функция Бесселя
У нас есть следующее рекурсивное отношение между модифицированными функциями Бесселя первого рода и порядком $n$, $I_n(x)$

\begin{align*}
x\frac{dI_n(x)}{dx} + nI_n(x) &= xI_{n-1}(x) \\
\frac{dI_0(x)}{dx} &= I_1(x).
\end{align*}

Кроме того, у нас есть представления рядов $I_0(x)$ и $I_1(x)$

\begin{align*}
I_0(x) &= \sum\limits_{k=0}^{\infty} \frac{\left(\frac{x}{2}\right)^{2k}}{(k!)^2} \\
I_1(x) &= \sum\limits_{k=0}^{\infty} \frac{\left(\frac{x}{2}\right)^{2k+1}}{(k!)(k+1)!}.
\end{align*}
