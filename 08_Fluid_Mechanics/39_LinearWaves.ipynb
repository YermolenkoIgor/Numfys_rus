На этом уроке мы представим некоторые методы анализа линейных волн в бесконечных областях. В конечных областях *разделение переменных* позволяет нам определять естественные режимы колебаний поверхности свободной жидкости. Из-за линейности управляющих уравнений любое решение может быть записано как суперпозиция таких режимов. Однако в бесконечной области общий поток может быть записан как суперпозиция гармонических волн. Это создает идеальную среду для использования *преобразования Фурье*. 

В частности, мы рассмотрим задачу нахождения временной эволюции начального возмущения $\eta_0(x)$ поверхности невязкой и несжимаемой жидкости на постоянной глубине.

## Свойства преобразования Фурье 

Мы кратко представим некоторые важные свойства преобразования Фурье, которые будут полезны в этой задаче. 

Пусть $f$ - достаточно хорошо управляемая функция. Мы определяем преобразование Фурье $\widehat{f}$ с помощью 

\begin{equation}
\widehat{f} (k) = \frac{1}{\sqrt{2\pi}} \int_{-\infty}^{\infty} f(x) e^{-ikx} \, \text{d}x.
\end{equation}

### Обратное преобразование Фурье

Исходная функция $f$ получается с помощью обратного преобразования, определенного

\begin{equation}
f(x) = \frac{1}{\sqrt{2\pi}} \int_{-\infty}^{\infty} \widehat{f}(k) e^{ikx} \, \text{d}x.
\end{equation}

### Преобразование Фурье производных

Как и в случае преобразования Лапласа, преобразование Фурье превращает обыкновенные дифференциальные уравнения в алгебраические уравнения, поскольку производные по $x$ преобразуются в величины кратные $ik$: 

\begin{equation}
\widehat{\frac{\text{d}^n f}{\text{d}x^n}} = \left(ik\right)^n \widehat{f}.
\end{equation}

## Управляющие уравнения для волн Стокса 

Для гравитационных волн малой амплитуды в несжимаемой жидкости у нас есть следующие управляющие уравнения для потенциала скорости $\phi(x,y,t)$ и смещения свободной поверхности $y = \eta(x,t)$ (см. Приложение).

\begin{align}
	\frac{\partial^2\phi}{\partial x^2} + \frac{\partial^2\phi}{\partial y^2} = 0& \quad \quad ; \quad \quad y < 0 \\
	\frac{\partial\phi}{\partial y} = \frac{\partial \eta}{\partial t}, \quad \frac{\partial \phi}{\partial t} + g\eta = 0& \quad \quad ;\quad \quad y = 0 \\
	\frac{\partial \phi}{\partial y} = 0& \quad \quad ; \quad \quad y = -h
\end{align}

где $h$ - средняя глубина воды, а $g$ - ускорение силы тяжести. Первое уравнение представляет собой непрерывное уравнение для несжимаемой жидкости, а именно $\mathbf{\nabla} \cdot \mathbf{v} = 0$, что переводится в $\nabla^2\phi = 0$ для потенциала скорости в случае безоборотного потока. Два следующих уравнения являются кинематическими и динамическими граничными условиями, а последнее - непроницаемость внизу. Мы определяем начальные условия поверхности свободной жидкости и для простоты предполагаем, что поверхность изначально находится в состоянии покоя

$$
	\eta(x,0) = \eta_0(x), \quad \frac{\partial\eta}{\partial t} = 0.
$$

Мы берем преобразование Фурье приведенных выше уравнений относительно $x$. Это приводит к следующему

\begin{align}
\frac{\partial^2 \widehat{\phi}}{\partial y^2} - k^2 \widehat{\phi} = 0& \quad \quad ; \quad \quad y < 0 \\
\frac{\partial \widehat{\phi}}{\partial y} = \frac{\partial \widehat{\eta}}{\partial t}, \quad \frac{\partial \widehat{\phi}}{\partial t} + g\widehat{\eta} = 0& \quad \quad ;\quad \quad y = 0 \\
\frac{\partial \widehat{\phi}}{\partial y} = 0& \quad \quad ; \quad \quad y = -h \\
\widehat{\eta} = \widehat{\eta_0}, \quad \frac{\partial \widehat{\eta}}{\partial t} = 0& \quad \quad ; \quad \quad t = 0 
\end{align}

Решения первого уравнения представляют собой линейную комбинацию экспонент, которая вместе с условием непроницаемости в третьем уравнении дает

\begin{equation}
	\widehat{\phi} = A(k,t) \cosh{ \left(k ( y + h)\right)},
\end{equation}

где амплитуда может зависеть от $k$ и $t$. Из второго уравнения следует, что амплитуда удовлетворяет

\begin{equation}\label{eq:a1}
	Ak \sinh{kh}  = \frac{\partial \widehat{\eta}}{\partial t} \quad (1)
\end{equation} 

и

\begin{equation}\label{eq:a2}
	\frac{\partial A}{\partial t}\cosh{kh} + g\widehat{\eta} = 0.  \quad (2)
\end{equation}

Дифференцируя уравнение (1) по времени, мы получаем $\frac{\partial A}{\partial t} k \sinh{kh} = \frac{\partial^2\widehat{\eta}}{\partial t^2}$, что мы можем подставить в уравнение (2), чтобы получить

$$
	\frac{\partial^2\widehat{\eta}}{\partial t^2} + gk \tanh{kh} \cdot \widehat{\eta}  = 0.
$$

Это уравнение простого гармонического движения. Применяя начальные условия, мы получаем выражение для $\widehat{\eta}$

\begin{equation}
	\widehat{\eta} = \widehat{\eta_0} \cos{\omega t}
\end{equation}

где $$\omega(k) = \sqrt{gk \tanh{kh}}$$ - дисперсионное соотношение. Временная эволюция свободной поверхности жидкости получается обратным преобразованием Фурье,

\begin{equation}\label{eq:eta}
	\eta(x,t) = \frac{1}{\sqrt{2\pi}} \int_{-\infty}^{\infty} \widehat{\eta_0}(k) \cos{\left(\omega(k) t\right)} e^{ikx} \, \text{d}k. \quad (3)
\end{equation}

К сожалению, найти точное и явное выражение для $\eta(x,t)$ сложно даже для простых дисперсионных соотношений $\omega(k)$. На самом деле, численное решение этой задачи также не совсем просто, поскольку оно, по сути, включает в себя решение двух интегралов по бесконечным областям. 
Нам нужно решить как интеграл в уравнении (3), так и интеграл, определяющий $\widehat{\eta_0}(k)$.
С этой целью мы представим некоторые подходы к получению $\eta(x, t)$ численно и аналитически. Мы ограничим наше внимание гауссовским начальным возмущением $\eta_0(x) = a \exp{\left(- x^2 /2L^2 \right)}$.

## Гауссово возмущение

Для начальной формы свободной поверхности жидкости, равной $\eta_0(x) = a \exp\left(-x^2/2L^2\right)$, мы решаем только *один* из интегралов, так как $\widehat{\eta_0}$ может быть вычислено 

\begin{equation}
	\widehat{\eta_0}(k) = \frac{1}{\sqrt{2\pi}} \int_{-\infty}^{\infty} a \exp{\left(-\frac{x^2}{2L^2}\right)} \exp{(ikx)} \, \text{d} x = aL \exp{\left(-\frac{k^2L^2}{2}\right)}.
\end{equation}

Если мы вставим это в уравнение (3), мы получим

\begin{equation}\label{eq:eta2}
	\eta(x,t) = \int_{-\infty}^{\infty} \frac{\text{d}k}{\sqrt{2\pi}} aL\exp{\left(-\frac{k^2L^2}{2}\right)} \cos{\omega(k) t}\exp{(ikx)}.
\end{equation}

При работе с этим выражением полезно вводить безразмерные величины для представления $x$ и $t$. Естественная шкала времени этой системы составляет $\sqrt{L/g}$, а полезная шкала длины - $\sqrt{2}L$. Это позволяет нам определить следующие безразмерные величины 

\begin{equation}
\xi = \frac{x}{\sqrt{2}L} \quad ; \quad \tau = \frac{t}{\sqrt{L/g}}
\end{equation}

```python
# Importing useful packages

import numpy as np 
import matplotlib.pyplot as plt 
import matplotlib as mpl
from matplotlib import rc
import matplotlib.gridspec as gridspec

eps = np.finfo(float).eps  # Machine decimal rounding

# Setting common plotting parameters
fontsize = 22
newparams = {'axes.titlesize': fontsize, 'axes.labelsize': fontsize,
             'lines.linewidth': 2, 'lines.markersize': 7,
             'figure.figsize': (13, 7), 'ytick.labelsize': fontsize,
             'xtick.labelsize': fontsize, 'legend.fontsize': fontsize,
             'legend.handlelength': 1.5, 'figure.titlesize': fontsize,
             'figure.dpi':200, 'text.usetex': True, 'font.family': 'sans-serif'}
plt.rcParams.update(newparams)
```

```python
# Physical constants 

g = 10    # Acceleration of gravity

global h  # Making h global so we can change it afterwards easily
h = 10    # Depth of water

a = 1     # Initial height of gaussian perturbation

global L  # Making L global so we can change it afterwards easily
L = 0.5   # Initial width of gaussian

t_0 = np.sqrt(L/g)  # Time scale
x_0 = np.sqrt(2)*L  # Length scale in x-direction
```

```python
def plot_waves(X,Y,title,labels):
    """
    Function for plotting waves.
    
    Args:
        X (np.array)      : x-values for wave
        Y (np.array)      : y-values for wave. Each row represents one wave.
        title (string)    : title of plot
        labels (np.array) : array of labels corresponding to each wave
    """
    
    fig = plt.figure()
    plt.title(title)
    
    for i in range(Y.shape[0]):
        plt.plot(X,Y[i,:],label = labels[i])
    
    plt.xlabel(r"$\xi = \frac{x}{\sqrt{2}L}$")
    plt.ylabel(r"$y / a$")

    plt.grid(linestyle="--")

    plt.tight_layout()
    plt.legend()
```

```python
def f(xi):
    """
    Initial gaussian disturbance.
    
    """
    return np.exp(-xi**2)

X = np.linspace(-20,20,500)
plot_waves(X,np.array([f(X)]),title = r"Initial shape of free fluid surface", labels = [r"$\eta_0(\xi)$"])
```

## Изменение границ интеграла

Поскольку с бесконечной областью интегрирования трудно иметь дело численно, мы могли бы попытаться заменить $\infty$ в границах на какое-то большое число, скажем, $M$. В конце концов, когда мы имеем дело с резко локализованной начальной волной, подынтегральное выражение должно довольно быстро затухать, и поэтому ошибка, вызванная введением такого ограничения, должна быть небольшой, когда волновой пакет достаточно локализован (т.е. через короткое время после того, как произошло возмущение). В этом контексте мы решили использовать $M = 50$ с разбиением по 500 в приближении трапеции. 
Для получения дополнительной информации об этом алгоритме обратитесь к этому [блокноту](https://nbviewer.jupyter.org/urls/www.numfys.net/media/notebooks/numerical_integration.ipynb). 


```python
global M
M = 50
```

```python
# The decorator @np.vectorize vectorizes the function so that it can take as 
# argument a nested sequence of objects or numpy arrays as inputs and returns a single numpy array
# for the documentation, see: https://numpy.org/doc/stable/reference/generated/numpy.vectorize.html

@np.vectorize
def comp_trap(f,a,b,N = 500):
    """
    Simple implementation of the trapezoid rule.
    
    Args:
        f (function) : integrand
        a (float)    : start point
        b (float)    : end point
        N (int)      : number of intervals, default = 500
    
    Returns:
        I (float)    : trapezoid integral approximation
        
    """
    I = 0
    delta_x = (b-a)/N
    for i in range(N):
        I += 0.5*delta_x*(f(a+delta_x*i)+f(a+delta_x*(i+1))) 
    return I
```

```python
def omega_grav(k):
    """
    Function for gravity wave dispersion relation, omega(k)
    
    """
    return np.sqrt(g*k*np.tanh(k*h))
```

```python
def eta_crude(xi,tau,N = 500,omega = omega_grav):
    """
    Function for determining eta(xi,tau) using the crude 
    trapezoid approximation with the upper bounds changed from infinity to M 
    
    Args:
        xi (float)       : xi-value(s)
        tau (float)      : tau-value 
        N (int)          : number of intervals to use in trapezoid approx, default = 500
        omega (function) : dispersion relation, default = omega_grav
        
    Returns:
        eta(xi,tau)
    """
    prefac = L /(np.sqrt(2*np.pi))
    
    def integrand(k):
        return prefac * np.exp(-k**2*L**2/2)*np.cos(omega(k)*tau*t_0)*np.exp(1j*k*xi*x_0)
    
    return comp_trap(integrand,-M,M,N = N)
```

```python
Y1 = eta_crude(X,0)
Y2 = eta_crude(X,3)
Y3 = eta_crude(X,7)

Y = np.array([np.real(Y1),np.real(Y2),np.real(Y3)])

plot_waves(X, Y,title = r"Time evolution of free fluid surface",
           labels = [r"$\eta(\xi,\tau = 0)$",r"$\eta(\xi,\tau = 3)$",r"$\eta(\xi,\tau = 7)$"])
```

Чтобы сэкономить некоторое время, мы включаем один метод грубого приближения трапеции, используя больше функций, предлагаемых $\texttt{NumPy}$. Чтобы написать функцию `comp_trap`, используя подход $\texttt{Numpy}$, можно сделать что-то похожее на то, что показано в функции `comp_trap_numpy` в ячейке ниже. Однако, поскольку мы хотим иметь возможность вызывать функцию $\eta$ в массиве $\xi$-значений, нам придется переписать всю функцию, поскольку бродкаст-функции  $\texttt{NumPy}$ не делают того, что мы намереваемся в этом контексте. Естественная структура для решения этой проблемы в $\texttt{NumPy}$ - это сетки, см., например [этот блокнот](https://nbviewer.jupyter.org/urls/www.numfys.net/media/notebooks/NumpyIntermediate.ipynb#Meshgrid.) для подробного обзора. Приведенный ниже простой тест действительно подтверждает, что этот метод быстрее, и поэтому мы будем использовать этот метод при работе с составленным правилом трапеции.

```python
def comp_trap_numpy(f, a, b, N=500):
    """This function is not used! 
    It is an example of how one could write comp_trap using NumPy.
    However, simply swapping comp_trap with comp_trap_numpy in eta_crude
    does not work, as NumPy broadcasting does not do what we intend it to in this context.
    Thus, one has to rewrite eta_crude as well.
    """
    x, dx = np.linspace(a, b, N, retstep=True)
    I = dx * (np.sum(f(x)) - 0.5*f(x[0]) - 0.5*f(x[-1]))
    return I
```

```python
def eta_crude_numpy(xi,tau, N = 500, omega = omega_grav):
    """
    Function for determining eta(xi,tau) using the crude 
    trapezoid approximation with the upper bounds changed from infinity to M,
    using more of the functionality provided by NumPy.
    
    Args:
        xi (float)       : xi-value(s)
        tau (float)      : tau-value 
        N (int)          : number of intervals to use in trapezoid approx, default = 500
        omega (function) : dispersion relation, default = omega_grav
        
    Returns:
        eta(xi,tau)
    """
    prefac = L /(np.sqrt(2*np.pi))
    
    def integrand(k):
        return prefac * np.exp(-k**2*L**2/2)*np.cos(omega(k)*tau*t_0)*np.exp(1j*k*xi*x_0)
    
    k, dk = np.linspace(-M,M,N,retstep = True)
    
    xixi, kk = np.meshgrid(xi,k)
    integrand_mesh = integrand(kk)
    
    I = dk * (np.sum(integrand_mesh, axis=0) - 0.5*integrand_mesh[0, :] - 0.5*integrand_mesh[-1, :])
    
    return I
```

## Преобразование интеграла в конечную область

Другой подход к работе с бесконечной областью состоит в том, чтобы ввести изменение переменных, скажем, $\zeta(k)$ так, чтобы $\zeta(\pm\infty)$ было конечным. Есть много вариантов для такого сопоставления, но давайте для начала попробуем $\zeta (k) = \tanh(k)$. Тогда $\text{d}\zeta = (1- \tanh^2k)\,\text{d}k = (1-\zeta^2)\,\text{d}k$ и $\zeta (k = \pm\infty) = \pm 1$. Это превращает выражение для свободной поверхности жидкости в момент времени $t$ в 

\begin{equation*}
	\eta(x,t) = \int_{-1}^{1} \frac{\text{d}\zeta}{\sqrt{2\pi}} \frac{aL}{1-\zeta^2} \exp{\left(- \frac{\text{artanh}^2(\zeta)L^2}{2}\right)} \cos{\left(\omega(\text{artanh }(\zeta))t\right)} \exp{\left(i\text{artanh}(\zeta)x\right)}.
\end{equation*}

При выполнении этого интеграла численно очевидно, что нам придется избегать $\zeta = \pm 1$. 
Мы достигаем этого, изменяя интервал интегрирования на $\zeta \in [-1 + 2 \epsilon, 1 - 2\epsilon]$, где $\epsilon$ это машинная эпсилон.

```python
def eta_exact(xi,tau,N = 500, omega = omega_grav):
    """
    Function for determining eta(xi,tau) using the 'exact' method of  
    mapping the coordinates to [-1,1], and then using trapezoid method to calculate integral
    
    Args:
        xi (float)       : xi-value(s)
        tau (float)      : tau-value 
        N (int)          : number of intervals to use in trapezoid approx, default = 500
        omega (function) : dispersion relation, default = omega_grav
        
    Returns:
        eta(xi,tau)
    """
    def integrand(zeta):
        return L/(np.sqrt(2*np.pi)) * 1/(1-zeta**2) * np.cos(omega(np.arctanh(zeta))*tau*t_0) * np.exp(- np.arctanh(zeta)**2 * L**2 /2 + 1j * xi * x_0 * np.arctanh(zeta)) 
    
    # We add machine epsilon, `eps`, to avoid the endpoints xi = +-1.
    k, dk = np.linspace(-1 + 2*eps,1-2*eps,N,retstep = True)
    
    xixi, kk = np.meshgrid(xi,k)
    integrand_mesh = integrand(kk)
    
    I = dk * (np.sum(integrand_mesh, axis=0) - 0.5*integrand_mesh[0, :] - 0.5*integrand_mesh[-1, :])
    
    return I
```

```python
Y1 = eta_exact(X,0)
Y2 = eta_exact(X,3)
Y3 = eta_exact(X,7)

Ys = np.array([np.real(Y1),np.real(Y2),np.real(Y3)])

plot_waves(X, Ys,title = r"Time evolution of free fluid surface",
           labels = [r"$\eta(\xi,\tau = 0)$",r"$\eta(\xi,\tau = 3)$",r"$\eta(\xi,\tau = 7)$"])
```

При использовании точно такого же наивного правила трапеции для этого интеграла мы ясно видим, что это приводит к очень неудовлетворительным результатам. Для небольших $\xi$ результаты в порядке, но становятся хуже по мере увеличения $\xi$. Построив подынтегральное выражение (скажем) $\tau = 3$ для некоторых разных $\xi$, мы можем получить представление о том, чем это вызвано.

```python
def integrand(zeta,xi,tau):
    return L/(2*np.pi) * 1/(1-zeta**2) * np.cos(omega_grav(np.arctanh(zeta))*tau*t_0) * np.exp(- np.arctanh(zeta)**2 * L**2 /2 + 1j * xi * x_0 * np.arctanh(zeta))
    
Z1 = np.linspace(0.95,1 - eps,500)    # array of zeta values between 0.95 and 1-eps 
Z2 = np.linspace(-1+ eps,1 - eps,500) # array of zeta values between -1 + eps and 1 + eps

# Integrand evaluated at different \xi values on the short interval

I1_short = integrand(Z1,0,3)
I2_short = integrand(Z1,1,3)
I3_short = integrand(Z1,5,3)
I4_short = integrand(Z1,10,3)

# Integrand evaluated at different \xi values on the large interval

I1 = integrand(Z2,0,3)
I2 = integrand(Z2,1,3)
I3 = integrand(Z2,5,3)
I4 = integrand(Z2,10,3)

fig = plt.figure(figsize=(13,10))

fig.suptitle(r"Integrand at $\tau = 3$")

gs = gridspec.GridSpec(2,1) # Making a 2 x 1 grid

ax = plt.subplot(gs[0,:])   # First axis is the first row

ax.plot(Z2,np.real(I1),label = r"$\xi = 0$")
ax.plot(Z2,np.real(I2),label = r"$\xi = 1$")
ax.plot(Z2,np.real(I3),label = r"$\xi = 5$")
ax.plot(Z2,np.real(I4),label = r"$\xi = 10$")

plt.xlabel(r"$k$")
plt.grid(linestyle="--")

# Placing legend outside plot
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5),
          ncol=1, fancybox=True, shadow=True)

ax = plt.subplot(gs[1,:])    # Second axis is the second row

ax.plot(Z1,np.real(I1_short),label = r"$\xi = 0$")
ax.plot(Z1,np.real(I2_short),label = r"$\xi = 1$")
ax.plot(Z1,np.real(I3_short),label = r"$\xi = 5$")
ax.plot(Z1,np.real(I4_short),label = r"$\xi = 10$")

# Placing legend outside plot
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5),
          ncol=1, fancybox=True, shadow=True)

plt.xlabel(r"$k$")
plt.grid(linestyle="--")

plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # tight_layout adjusted to work with suptitle
```

Подынтегральное выражение очень быстро изменяется вблизи $\pm 1$ для увеличения $x$, и поэтому становится трудно точно интегрировать его с равноудаленными абсциссами интеграции. Мы можем попытаться решить эту проблему, не используя равномерно распределенные точки интеграции, а скорее пусть больше из них будет расположено вблизи $\pm 1$. В качестве альтернативы мы можем более разумно выбрать наше отображение координат, что позволит нам преобразовать задачу в вычисления ряда Фурье.

### Квадратура Фурье - Чебышева

Другим способом отображения координат на конечный интервал является рассмотрение преобразования формы $k = \chi(\theta) =  J \cot{\theta}$, эффективно преобразуя наш интервал интегрирования в $[0,\pi]$. Это видно из того факта, что $\chi'(\theta) = -J \frac{1}{\sin^2(\theta)}$, и что $\chi(\theta = \pi) = -\infty$ и $\chi(\theta = 0) = \infty$. Из этого следует, что 
\begin{equation*}
	\eta(\xi,\tau) = \int_{-\infty}^{\infty} f (k;\xi,\tau)\, \text{d}k = J \int_{0}^{\pi} \text{d}\theta \frac{f(J \cot{\theta};\xi,\tau)}{\sin^2{\theta}},
\end{equation*}

где $J$ - определенная пользователем константа. Одно из преимуществ, связанных с отображением на $[0,\pi]$, заключается в том, что легко вычислить интеграл синусов и косинусов на этом интервале. Поэтому мы пытаемся развернуть подынтегральное выражение в его ряд Фурье, а затем интегрировать его. Идея заключается в следующем (см., например. [[1]](#fourier_quad)):

Мы хотим вычислить интеграл 

\begin{equation}\label{eq:int}
I = \int_{0}^{\pi} f[\chi(\theta)] \chi'(\theta) \, \text{d}\theta, \quad (4)
\end{equation}

где $\chi$- функция, которая отображает наш интервал $[- \infty,\infty]$ в $[0,\pi]$. В нашем случае мы выбираем $\chi(\theta) = J \cot{\theta}$. Обратите внимание, что, поскольку эта функция фактически отображает верхнюю границу ($+ \infty$) в $0$, а нижнюю границу в $\pi$, но, кроме того, содержит знак минус в производной, приведенное выше обозначение $\chi'(\theta)$ несколько оскорбительно, поскольку оно игнорирует знак минус в производной.

Предположим, что мы приближаем подынтегральное выражение $f(\chi(\theta))$ его усеченным разложением в ряд Фурье с помощью тригонометрической интерполяции, то есть 

\begin{equation}\label{eq:trig_interpolate}
q(\theta) = a_0 + \sum_{n=1}^N a_n \cos{n \theta} + \sum_{n=1}^{N} b_n \sin{n \theta}, \quad (5)
\end{equation}

где $N$-некоторая правильно выбранная точка отсечения.

Теперь предположим, что мы объединяем тригонометрические базисные функции в новые кардинальные функции $C_j$ для каждой точки интерполяции $t_j$ со свойством

\begin{equation}\label{eq:cardinal_funcs}
C_j(t_i) = \begin{cases} 
1, i = j\\
0, i \neq j
\end{cases}. \quad (6)
\end{equation}

Явную форму этих кардинальных функций можно найти в приложении.
При вставке в уравнение (5) это дает

$$
q(\theta) = \sum_{j=0}^{N-1} q(t_j) C_j(\theta).
$$

Последним шагом является интеграция $q(\theta)$ для аппроксимации интересующего нас интеграла в (4),

$$
I \approx \sum_{j=0}^{N-1} w_j f (\chi(t_j)),
$$
где веса $w_j$ определяются как
$$
w_j = \chi'(t_j) \int_{0}^{\pi} C_j(\theta)\, \text{d}\theta.
$$

При использовании сопоставления $k \mapsto J \cot{\theta}$ можно вычислить веса, $w_j$ и аппроксимацию для точек интерполяции $N$.

\begin{equation*}
    \eta(\xi,\tau) \approx \frac{J\pi}{N-1} \sum_{n = 1}^{N-2} \frac{f(J \cot{(\pi n/(N-1))};\xi,\tau)}{\sin^2{(n\pi/(N-1))}} + \frac{J\pi}{2N-2} \left( \frac{f(J \cot{(0)};\xi,\tau)}{\sin^2{(0)}} + \frac{f(J \cot{(\pi)};\xi,\tau)}{\sin^2{(\pi)}} \right),
\end{equation*}

где мы предполагаем, что два последних слагаемых конечны. Вывод этих формул можно найти в статье, упомянутой в [[1]](#fourier_quad), и некоторые подробности представлены в приложении. Чтобы избежать деления на $0$, мы оцениваем $\sin$ и $\cot$ в $\epsilon$ и $\pi - \epsilon$ вместо $0$ и $\pi$, где $\epsilon$ - это машинное ограничение для десятичного округления.

```python
def eta_exact_good(xi,tau,J = 10 ,N = 500,omega = omega_grav):
    """
    Function for determining eta(xi,tau) using the Fourier chebyshev method,
    with coordinate mapping k = J cot (theta) in the integral.
    
    Args:
        xi (float)       : xi-value(s)
        tau (float)      : tau-value 
        n (int)          : number of intervals to use in trapezoid approx, default = 500
        omega (function) : dispersion relation, default = omega_grav
        
    Returns:
        eta(xi,tau)
    """
    prefac = L /(np.sqrt(2*np.pi))
    
    def integrand(k):
        return prefac * np.exp(-k**2*L**2/2)*np.cos(omega(k)*tau*t_0)*np.exp(1j*k*xi*x_0)
    
    S1 = sum(J*np.pi/(N-1) * integrand(J*1/np.tan(n*np.pi/(N-1)))*1/(np.sin(n*np.pi/(N-1))**2) for n in range(1,N-1))
    S2 = J*np.pi/(2*N-2)*integrand(J*1/np.tan(eps))*1/((np.sin(eps))**2)
    S3 = J*np.pi/(2*N-2)*integrand(J*1/np.tan(np.pi - eps))*1/((np.sin(np.pi - eps))**2)
    
    return S1 + S2 + S3
    
```

```python
Y1 = eta_exact_good(X,0)
Y2 = eta_exact_good(X,3)
Y3 = eta_exact_good(X,7)

Ys = np.array([np.real(Y1),np.real(Y2),np.real(Y3)])

plot_waves(X,Ys,title = r"Time evolution of free fluid surface",
           labels = [r"$\eta(\xi,\tau = 0)$",r"$\eta(\xi,\tau = 3)$",r"$\eta(\xi,\tau = 7)$"])
```

Судя по графику, результаты преемлемы для выбранной области. Определяемая пользователем константа $J$ в отображении координат, конечно, может варьироваться. На самом деле оптимальное значение $J$ зависит от задачи и может быть аппроксимировано обычными методами оптимизации. Ниже мы представляем простой метод выбора наилучшего $J$ из заданного массива значений.

В этом методе мы используем приближение трапеции с большим количеством панелей в качестве "правильного" опорного значения для сравнения метода Фурье-Чебышева. По сути, мы ищем значение $J$, для которого метод Фурье-Чебышева наиболее близок к этому эталонному значению. 


```python
def search_optimal_J(Js,Xis,tau,omega = omega_grav):
    """
    Function for finding the J-value in a given array
    which gives the smallest deviation between the 'brute-force'
    method of calculating eta, and the Fourier chebychev integration.
    
    Args:
        Js (np.array)     : array of J-values to search through
        Xis (np.array)    : the xi-values to evaluate eta over
        tau (float)       : the time to evaluate eta at
        omega (function)  : dispersion relation, default = gravity wave
        
    Returns:
        best_J (float)    : J giving smallest deviation
        err (float)       : smallest deviation obtained
        J_stats (np.array): 2d array with all the Js and their corresponding error
        
    """
    
    
    correct_eta = eta_crude_numpy(Xis,tau,N=5000,omega=omega_grav)
    
    best_J = Js[0]
    eta_0 = eta_exact_good(Xis,tau,J = best_J, N = 500, omega = omega_grav)
    err = np.linalg.norm(correct_eta - eta_0)
    
    n = Js.size
    J_stats = np.zeros((n,2))
    J_stats[:,0] = Js
    
    for index, J in enumerate(Js):
        new_eta = eta_exact_good(Xis,tau,J=J,N = 500, omega = omega_grav)
        new_err = np.linalg.norm(new_eta - correct_eta)
        if new_err < err:
            err = new_err
            best_J = J
            
        J_stats[index,1] = new_err
            
    return best_J, err, J_stats
```

```python
J = np.linspace(0.01,100,500)
best_J,best_diff, stats = search_optimal_J(J,X,3)
```

```python
print("Best J: ", best_J)
print("Error at best J: ", best_diff)
```

Поскольку мы видим, что выбор подходящего значения $J$ делает отклонение от эталонного значения очень небольшим, метод Фурье-Чебышева, очевидно, намного лучше, чем наивный метод трапеций в этом контексте. Чтобы увидеть это в явном виде, сравните время, затраченное на грубое приближение с 5000 шагами, и интеграл Фурье-Чебышева с 500 точками интерполяции:

```python
%timeit eta_crude_numpy(X,0,N=5000,omega=omega_grav)
%timeit eta_exact_good(X,0,J = best_J,N = 500,omega=omega_grav)
```

Мы включаем график ошибки в зависимости от константы $J$ в отображение координат, показывая, что выбор правильного $J$ на самом деле является критическим для того, чтобы этот метод был лучше, чем обычная трапеция.

```python
plt.title(r"Error as a function of $J$ in the coordinate mapping")

plt.plot(stats[:,0],stats[:,1],label=r"\texttt{err}($J$)")
plt.plot(best_J,best_diff,"ro",label=r"$\min$ \texttt{err}")

plt.yscale("log")
plt.xscale("log")

plt.ylabel(r"\texttt{err}")
plt.xlabel(r"$J$")

plt.grid(True, which="both", ls="--")

plt.legend()
plt.tight_layout()
```

## Аналитическое выражение для коротких промежутков времени 

Рассматривая короткие промежутки времени после начального возмущения, мы фактически можем получить аналитическое выражение для временной эволюции поверхности свободной жидкости. Нашей отправной точкой является общее выражение 

\begin{equation}
\eta(x,t) = \int_{-\infty}^{\infty} \frac{dk}{\sqrt{2\pi}} aL\exp{\left(-\frac{k^2L^2}{2}\right)} \cos{\omega(k) t}\exp{(ikx)}.
\end{equation}

Мы начинаем с предположения, что вода *глубокая*, то есть $\omega(k) = \sqrt{g |k|}$. В этом случае у нас есть

$$
\cos{\omega(k) t } = 1 - \frac{1}{2}  g|k| t^2 + \mathcal{O}(t^4).
$$

Для рассматриваемого возмущения мы будем иметь незначительный вклад в интеграл всякий раз, когда $|k| \gtrsim \sqrt{2}/L$, так как тогда амплитуда будет настолько мала, что она не будет иметь никакого отношения к значению интеграла. Кроме того, если мы рассмотрим времена, когда $\tau = t \sqrt{g/L} \ll 1$, мы знаем, что

$$
\sqrt{g|k|}t = \tau \sqrt{l|k|} \lesssim \sqrt{2} \tau \ll 1,
$$

следовательно, мы должны быть в состоянии сохранить только первые два члена в разложении Тейлора для $\cos$, не внося существенной ошибки в течение короткого времени. В этом случае у нас будет

\begin{align*}
\eta(x,t) &\approx \int_{-\infty}^{\infty} \frac{\text{d}k}{\sqrt{2\pi}} aL\exp{\left(-\frac{k^2L^2}{2}\right)}\exp{(ikx)} \left( 1 - \frac{1}{2} g |k|t^2 \right) \\
&= a  \exp{\left(-\frac{x^2}{2L^2}\right)} - \frac{aLgt^2}{2\sqrt{2\pi}} \int_{-\infty}^{\infty} \text{d}k |k| \exp{\left(-\frac{k^2L^2}{2}\right)}\exp{(ikx)} \\
\eta(\xi,\tau)&\approx a \bigg( \exp{\left(-\xi^2\right)} - \frac{\tau^2}{\sqrt{2\pi}} \left[ 1 - \sqrt{\pi} \xi \exp{\left(-\xi^2\right)} \text{erfi}(\xi)\right] \bigg),
\end{align*}

где мы выразили второй интеграл в терминах функции мнимой ошибки $\text{erfi}(\cdot)$.

```python
from scipy.special import erfi, dawsn

def eta_short_time(xi,tau):
    """
    Analytical expression for short-time approximation of eta(xi,tau)
    
    """
    
    return (np.exp(-xi**2) - tau**2*1/(np.sqrt(2*np.pi)) * (1 - np.sqrt(np.pi) * xi * np.exp(-xi**2) * erfi(xi)))

Y1 = eta_short_time(X,0)
Y2 = eta_short_time(X,0.1)
Y3 = eta_short_time(X,0.5)
Y4 = eta_short_time(X,1.5)

Ys = np.array([np.real(Y1),np.real(Y2),np.real(Y3),np.real(Y4)])

plot_waves(X, Ys, title = r"Time evolution of free fluid surface - short time approx.",
           labels = [r"$\eta(\xi,\tau = 0)$",r"$\eta(\xi,\tau = 0.1)$",r"$\eta(\xi,\tau = 0.5)$",r"$\eta(\xi,\tau = 1.5)$"])
```

```python
from scipy.stats import linregress
from mpltools import annotation  # The special package mpltools may need to be installed

def slope_log(X,Y):
    """
    Function for getting the approximate slope of the linear regression
    done on set of points (log(X),log(Y))
    
    Args:
        X (np.array) : x-values
        Y (np.array) : y-values
        
    Returns:
        slope (float) : a in expression log(y) = a * log(x) + b
    """
    slope, intercept, r_value, p_value, std_err = linregress(np.log(X),np.log(Y))
    return slope


def short_time_approx(X,T):
    """
    Function for comparing the short-time approximation with
    the 'exact' numerical plot. Plots the deviation between the two 
    at time-values in a given array.
    
    Args:
        T (np.array) : tau values to evaluate eta at 
        X (np.array) : xi-values to evaluate eta at
        
    """
    
    N_correct = np.array([eta_exact_good(X,t,J=best_J) for t in T])
    N_approx  = np.array([eta_short_time(X,t) for t in T])
    
    errs = np.array([np.linalg.norm(N_correct[i,:] - N_approx[i,:]) for i in range(len(T))])
    
    fig, ax = plt.subplots()
    
    plt.title(r"Error as function of $\tau$ from short time approximation")
    ax.plot(T,errs)
    
    plt.grid(ls="--")
    
    ax.set_xlabel(r"$\tau$")
    ax.set_ylabel(r"$\texttt{err}$")
    
    # Making a slope marker in the plot. More examples on how it's used can be found here:
    # http://tonysyu.github.io/mpltools/auto_examples/annotation/plot_slope_marker.html
    
    half_size = int(1/2 * np.size(errs)) # using only the upper half of the values
    
    slope = round(slope_log(T[half_size:],errs[half_size:]),3)
    
    ax.set_yscale("log")
    ax.set_xscale("log")
    
    annotation.slope_marker((1,1), (slope,1), invert=True, 
                        ax=ax,text_kwargs={"fontsize" : fontsize})
    
    plt.tight_layout()
    plt.show()
```

```python
T = np.logspace(-3,np.log(2),500)

short_time_approx(X,T)
```

Как и ожидалось, ошибка увеличивается с увеличением $\tau$. Более того, построив ее на $\texttt{log-log}$-графике, мы можем видеть, что у нас примерно есть ошибка $\mathcal{O}(\tau^4)$, чего мы и ожидали бы от нашего усечения $\cos$.

## Визуализация различных дисперсионных соотношений

Теперь, когда у нас есть метод определения временной эволюции поверхности свободной жидкости, было бы интересно визуализировать, как различные дисперсионные соотношения влияют на качественное поведение волны. Мы рассмотрим гравитационные волны, капиллярные волны и волны с линейной дисперсией

```python
# Function for animating waves in matplotlib: 
# Not relevant for anything else than visualisation purposes

import matplotlib
import matplotlib.animation
from IPython.display import HTML, clear_output

def animate_wave(x, wave0, fps,wave_func,title, t1=0, t2=None, realtime=True):
    assert not (not realtime and t2 is None), "non-realtime animation must be finite in time"
    
    dt = 1 / fps
    nframes = None if t2 is None else int((t2 - t1) / dt) # None animates forever
    
    nframesstr = "infinite" if t2 is None else f"{nframes}"
    durationstr = "infinite" if t2 is None else f"{t2-t1:.2f}"
    print("Animation information:")
    print(f"  Frames   : {nframesstr}")
    print(f"  Framerate: {fps} FPS")
    print(f"  Duration : {durationstr}")
    print(f"  Time step: {dt}")
    
    fig, ax = plt.subplots()
    
    ax.set_title(title)
    ax.set_xlabel(r"$\xi $")
    ax.set_ylabel(r"$\eta(\xi,t)$")
    ax.set_ylim(-1,1.5)
    ax.set_xlim(np.min(x),np.max(x))
    
    #plt.grid()
    ymax = max(np.max(np.abs(wave0)**2), np.max(np.abs(wave0)))
    graph2, = ax.plot([x[0], x[-1]], [0, -2*ymax]) # make room for negative values

    def animate(i):
        time = t1 + i*dt
        wave = wave_func(x, time)
        
        # set_data() etc. modifies existing an existing object in a figure
        # it is much more efficient than creating a new figure for every animation frame
        graph2.set_data(x, np.real(wave))
        graph2.set_color("blue")
        
        ax.collections.clear() # clears the fill_between from the previous instant
        ax.fill_between(x, -np.ones(np.shape(np.real(wave))), np.real(wave),color="blue",alpha=0.9)
           
    ani = matplotlib.animation.FuncAnimation(fig, animate, frames=nframes, interval=dt*1000, repeat=False)

    plt.close()
    
    if realtime:
        return ani
    else:
        return HTML(ani.to_jshtml())
```

```python
T   = 100   # Surface tension - choosing an artificially large number to see the effect of the capilliary forces
rho = 997   # Density of water 

def omega_cap(k):
    """
    Dispersion relation for capillary waves 
    
    """
    return np.sqrt((g*k + T*k**3/rho)*np.tanh(h*k))

def omega_linear(k):
    """
    Linear dispersion relation 
    
    """
    return np.sqrt(g*h) * k
```

```python
best_J_cap, _, _ = search_optimal_J(J,X,3,omega = omega_cap) # finding a good J for capillary waves 
best_J_lin, _, _ = search_optimal_J(J,X,3,omega = omega_linear) # finding a good J for linear dispersive waves 
```

```python
def eta_grav(xi,t):
    return eta_exact_good(xi,t/t_0,J = best_J ,N = 1000, omega = omega_grav)

def eta_cap(xi,t):
    return eta_exact_good(xi ,t/t_0,J = best_J_cap ,N = 1000, omega = omega_cap)

def eta_linear(xi,t):
    return eta_exact_good(xi,t/t_0,J = best_J_lin ,N = 1000, omega = omega_linear)
```

Давайте начнем с анимации простейшей волны, а именно волны с линейным дисперсионным соотношением, также известной как недисперсионные волны.

## Вывод управляющих уравнений для волн Стокса 

(Этот вывод является слегка адаптированной версией того, что содержится в [[2]](#linearwaves))

Для простоты мы рассмотрим случай двумерного потока. Пусть средняя глубина воды равна $h$, и предположим, что поверхность может быть записана как функция 

$$
y = \eta(x,t).
$$

В линейной волновой теории мы предполагаем, что волновое движение представляет собой небольшое возмущение, пропорциональное безразмерной величине $s = ka$, где $a$ - амплитуда, а $k$ - волновое число. Мы пренебрегаем всеми слагаемыми $\mathcal{O}(s^2)$.

В отсутствие волн скорость везде равна нулю, так что равновесное давление равно гидростатическому давлению $p_{eq}(y) = p_{\text{atm}} - \rho g y$.Таким образом, общее давление определяется 

$$
P(x,y,t) = p_{\text{atm}} - \rho g y + p(x,y,t),
$$

где $p$ - небольшое возмущение поля давления порядка $s$, обусловленное наличием волн.

Невязкое течение регулируется уравнениями Эйлера. В 2D они принимают форму,

\begin{align}\label{eq:pressure1}
	\frac{\partial u}{\partial t} + u \frac{\partial u}{\partial x} + v \frac{\partial u}{\partial y} &= - \frac{1}{\rho} \frac{\partial P}{\partial x} \quad (7)\\
	\label{eq:pressure2} 
	\frac{\partial v}{\partial t} + u \frac{\partial v}{\partial x} + v \frac{\partial v}{\partial y} &= - \frac{1}{\rho} \frac{\partial P}{\partial y} - g. \quad (8)
\end{align} 

Поле скоростей определяется $\mathbf{u}(x,y,t) = u(x,y,t) \mathbf{i} + v(x,y,t)\mathbf{j}$. Кроме того, мы предполагаем, что жидкость несжимаема, что дает уравнение непрерывности в виде 

\begin{equation}
	\frac{\partial u}{\partial x} + \frac{\partial v}{\partial y} = 0.
\end{equation}

Теперь $u, v, p$ имеют порядок $s$, поэтому мы можем игнорировать члены второго порядка в уравнениях (7) и (8), чтобы получить

\begin{align}
	\dot{u} &= - \frac{1}{\rho} \frac{\partial p}{\partial x} \\
	\dot{v} &= - \frac{1}{\rho} \frac{\partial p}{\partial y} 
\end{align}

Мы дифференцируем эти два уравнения относительно $y$ и $x$ соответственно, чтобы получить 

$$
-\frac{1}{\rho} \frac{\partial^2 p}{\partial x \partial y} = \frac{\partial \dot{u}}{\partial y} = \frac{\partial \dot{v}}{\partial x},
$$

или другими словами

$$
\frac{\partial}{\partial t} \left( \frac{\partial u}{\partial x} - \frac{\partial v}{\partial y} \right) = 0.
$$

Но мы распознаем величину в скобках как завихренность, поэтому это уравнение просто подразумевает, что завихренность везде постоянна во времени. Увидев волны раньше, мы немного знаем о типе ожидаемого решения, а именно о решениях, которые не являются стационарными, а периодическими во времени, и поэтому мы можем предположить, что если периодическое по времени решение имеет постоянную завихренность, а поскольку завихренности нет, когда амплитуда волны равна нулю (т.е. Неподвижная вода), это означает, что
скорее всего, завихренность фактически равна нулю при наличии волн. Действительно, в линеаризованном случае без фонового потока это оказывается правдой. В этом случае мы знаем, что можем несколько упростить процедуру, введя потенциал скорости; $\mathbf{u} = \nabla \phi$.

### Управляющее уравнение для $\phi$

Уравнение непрерывности для такого безтурбулентного потока дает управляющее уравнение для потенциала скорости,

$$
\nabla^2 \phi = \frac{\partial^2 \phi}{\partial x^2} + \frac{\partial^2 \phi}{\partial y^2} = 0.
$$

### Граничные условия

Простейшим граничным условием является то, что вода не может проходить через дно, т.е.,

$$
\frac{\partial \phi}{\partial y}\biggr\lvert_{y= - h} = 0,
$$

что называется условием непроницаемости.

Требуются еще два граничных условия, и оба они относятся к свободной поверхности. Чтобы получить то, что известно как *кинематическое граничное условие*, удобно определить функцию поверхности $F(x,y,t)$ как 

$$
F(x,y,t) = y - \eta(x,t).
$$

Сама поверхность неявно определяется $F(x,y,t) = 0$. Если в месте нахождения частицы жидкости это так (т.е. она на поверхности), это будет стационарным случаем. Это просто означает, что частица жидкости, которая в какой-то момент находится на свободной поверхности, всегда будет оставаться на поверхности, пока она движется вместе с потоком. Помните, что полная производная $\text{D}/\text{D}t = \partial/\partial t + \mathbf{u} \cdot \nabla$ - это изменение, наблюдаемое от частицы в жидкости, движущейся с потоком. Таким образом, с точки зрения частицы поверхности жидкости, $F$ не меняет своего значения с $0$,

\begin{equation}
    \frac{\text{D} F}{\text{D}t} = 0 \quad ; \quad y = \eta.
\end{equation}

Явно,

$$
\left( \frac{\partial}{\partial t} + u \frac{\partial}{\partial x} + v \frac{\partial }{\partial y} \right) \left[ y - \eta(x,t) \right] = - \dot{\eta} - u\frac{\partial \eta}{\partial x} + v = 0 \quad ; \quad y = \eta,
$$

что при линеаризации сводится к 

\begin{equation}\label{eq:kin_1}
    \dot{\eta} = v \quad ; \quad y = \eta.\quad (9)
\end{equation}  

Серьезное осложнение с граничным условием (9) остается, потому что оно оценивается в точке $y = \eta$, но $\eta(x, t)$ и само по себе неизвестно! Линеаризация позаботится об этом, потому что для линейного порядка не имеет значения, выполняется ли условие при $y = \eta$ или при $y = 0$. Поэтому мы получаем кинематическое граничное условие в виде 

\begin{equation}\label{eq:kin}
    \dot{\eta} = v = \frac{\partial \phi}{\partial y} \quad ; \quad y = 0.
\end{equation}

Конечное граничное условие называется динамическим граничным условием. Основная физика, лежащая в основе этого условия, заключается в признании того, что все напряжения в любой жидкой системе должны быть непрерывными функциями, как и на поверхности. Если бы силы, действующие на поверхностные частицы, не были непрерывными, конечные силы действовали бы на бесконечно малые жидкие элементы, приводя к бесконечному ускорению. В отсутствие вязкости, $\mu = 0$, непрерывность напряжения, нормального к поверхности, просто сводится к утверждению, что давление непосредственно над поверхностью должно равняться атмосферному давлению, а компоненты, параллельные поверхности, дают тривиальные уравнения, поскольку касательное напряжение имеет только вязкие компоненты.

В общем случае, когда поверхность изогнута, поверхностное натяжение вызовет ступенчатый разрыв давления по всей поверхности, но здесь мы это проигнорируем. Оставшееся динамическое граничное условие состоит просто в том, что давление должно быть непрерывным на границе раздела. Другими словами, давление непосредственно под поверхностью воды должно быть атмосферным. Теперь давление над поверхностью является постоянным (атмосферное давление), а давление под поверхностью описывается уравнением Бернулли, то есть 

$$
    \dot{\phi} + \nabla^2 \phi + gy + \frac{p}{\rho} = \text{constant},
$$

которое при линеаризации на поверхности жидкости сводится к 

$$
    \dot{\phi} + g\eta + \frac{p_{\text{atm}}}{\rho} = \text{constant} \quad ; \quad y = 0
$$

Поскольку оно должно выполняться и при спокойной воде $\phi = \eta = 0$, константа должна быть $p_{\text{atm}}/\rho$, и мы остаемся с *линеаризованным динамическим граничным условием*

\begin{equation}
    \dot{\phi} + g\eta = 0 \quad ; \quad y = 0.
\end{equation}

## Явное выражение для кардинальных функций $C_j(\theta)$

(Обратитесь к [[1]](#fourier_quad) для получения подробной информации по этой теме)

Комбинируя только косинусные функции (что нас интересует из-за области симметричного интегрирования), можно получить следующую кардинальную функцию для точек интерполяции $N$  $\left\lbrace t_i\right\rbrace_{i=0}^{N-1} = \left\lbrace \frac{i \pi}{N -1} \right\rbrace_{i = 0}^{N-1}$

\begin{align}
    C_j(\theta) &= \frac{2}{c_j (N-1)} \sum_{m=0}^{N-1} \cos{m t_j} \frac{\cos{m\theta}}{c_m} \\ 
                &= (-1)^{j+1} \sin{(\theta)} \sin{[(N-1)\theta ]} \frac{1}{(N-1) [ \cos{(\theta)} - \cos{(t_j)} ]}
\end{align}

где

$$
c_j = \begin{cases}
2, j = 0 \text{ or } N-1 \\
1, 1 < j < N-1
\end{cases}.
$$

Первая форма кардинальной функции дает понять, что на самом деле она представляет собой линейную комбинацию косинусов, в то время как вторая дает понять, что $C_j(t_i) = 0$, если $i\neq j$ и $1$, если $i = j$ (показывает случай, когда $i=j$ прямолинеен, но требует учитывать ограничение, например, с помощью правила Л'Опиталя). 

Мы не будем показывать эквивалентность двух формул для кардинальных функций, но мы вычисляем веса, используя первое представление.

\begin{align*}
w_j = \chi'(t_j) \int_{0}^{\pi} \text{d} \theta \frac{2}{c_j (N-1)} \sum_{m=0}^{N-1} \cos{m t_j} \frac{\cos{m\theta}}{c_m} 
    &= \chi'(t_j) \sum_{m=0}^{N-1} \frac{2}{c_jc_m (N-1)}  \cos{m t_j} \int_{0}^{\pi} \text{d}\theta \cos{(m\theta)} \\
    &= \chi'(t_j) \sum_{m=0}^{N-1} \frac{2}{c_jc_m (N-1)}  \cos{m t_j} \frac{\sin{m\pi}}{m} \\
    &= \chi'(t_j) \sum_{m=0}^{N-1} \frac{2}{c_jc_m (N-1)}  \cos{m t_j} \pi \delta_{m,0} \\
    &= \chi'(t_j) \frac{2\pi}{c_j c_0 (N-1)} \cos{(0)} = \chi' (t_j) \frac{\pi}{c_j (N-1)}
\end{align*}

Следовательно,

$$
w_j = \sin^{-2}\left({\frac{\pi j}{N-1}}\right) \cdot \begin{cases}
\frac{\pi}{N-1} , 0 < j < N-1 \\
\frac{\pi}{2(N-1)}, j = 0 \text{ or } j = N -1 
\end{cases}.
$$
