# Специальная теория относительности Эйнштейна и парадокс близнецов

# Замедление времени

В современной физике считается, что скорость света в вакууме одинакова, независимо от того, где она измеряется. Это заявление не кажется экстраординарным, пока не исследуются некоторые из последствий.

## Часы

Для дальнейших мысленных экспериментов, полагагаясь на инвариантность скорости света (обозначаемая символом $c$) представим часы - устройсво состоящее из излучателя света, зеркала и детектора - все находится внутри стеклянной вакуумной камеры.

Когда часы включаются, они испускают короткий импульс света, направленный на зеркало находящееся на расстоянии $d$ от источника. Импульс света проходит через камеру, отражается от зеркала и возвращается в исходное положение, где фиксируется детектором. При обнаружении светового импульса часы немедленно сообщают излучателю, чтобы он послал другой импульс, так что процесс циклически повторяется с периодом $T_{clock}$. 

Частоту, с которой "тикают" часы, легко вычислить:

$$f_{clock} = \frac 1 {T_{clock}} = \frac 1 {\frac {2d} c} = \frac c {2d}$$

Здесь начинаются странности: если поместить часы в быстро движущийся космический корабль и наблюдать за ними через окно из неподвижной точки, то должен проявиться интересный эффент: обитатели космического корабля испытывают течение времени медленнее, чем наблюдатель!

## Наблюдатель

Часы в космическом корабле ориентированы так, чтобы луч света указывал в направлении, перпендикулярном направлению движения корабля. Поэтому, при наблюдении за проносящимся мимо космическим кораблем можно заметить, как световой импульс на самом деле совершает небольшое зигзагообразное движение.

![img1](images/rel_1.png)

Поскольку луч света движется с постоянной скоростью $c$, вертикальная составляющая его движения, $c_y$ уменьшается, в то время как горизонтальная составляющая его движения, $c_x$, равна скорости космического корабля, $v$. Как сказал бы Пифагор:

$$c^2 = v^2 + c_y^2$$

и

$$c_y = \sqrt{c^2-v^2}$$

Поскольку время, необходимое для совершения одного тика, определяется через $c_y$ и $d$, можно сделать вывод, что частота часов космического корабля теперь:

$$f_{observed} = \frac 1 {T_{observed}} = \frac 1 {\frac {2d} {c_y}} = \frac {c_y} {2d} = \frac {\sqrt{c^2-v^2}} {2d}$$

Другой способ взглянуть на это - сравнить наблюдаемую (с земли) частоту с фактической (наблюдаемой на борту космического корабля) частотой часов:

$$\frac {f_{observed}} {f_{clock}} = \frac {\frac {\sqrt{c^2-v^2}} {2d}} {\frac c {2d}} = \frac {\sqrt{c^2-v^2}} {c} = \sqrt {1-\frac {v^2}{c^2}}$$


Этот эффект известен как релятивистское **замедление времени**.

## Странные последствия

Допустим, космический корабль движется со скоростью 8/10 скорости света. В этом случае мы можем легко вычислить, насколько медленнее тикают часы космического корабля, чем, скажем, эталонные часы рядом с наземным наблюдателем:

```python
from math import *
c = 3.00E8 # скорость света, метров в секунду
v = 0.8*c  # очень быстрый космический корабль
fr = sqrt(1 - v**2/c**2) # соотношение частот: наземный наблюдатель к наблюдателю в космическом корабле
print("отношение частоты у наземного наблюдателя к частоте у наблюдателя в космическом корабле: ", fr)
```

Это соотношение частот означает, что за каждые 100 минут, которые проходят для наземного наблюдателя, наблюдатель в космическом корабле, движущемся со скоростью 8/10 скорости света, будет видеть, как проходит ровно 60 минут. Если космический корабль находится в полете достаточно долго, разница в наблюдаемом времени составит года. Астронавт на космическом корабле буквально вернется домой, состарившись значительно меньше, чем люди, которые остались!

## И парадокс!

Погодите минутку! А нельзя ли сказать обратное? Разве астронавт не видит, что "наземный наблюдатель" тоже движется со скоростью 8/10 скорости света? Почему астронавт не обнаруживает, что у людей на земле замедлились часы?"

Это **парадокс близнецов**. Поскольку мы описали искажающий время эффект высокоскоростного перемещения, нет никаких предпосылок, чтобы этот эффект был не симметричен. Каждый участник (астронавт и наземный наблюдатель) видит, как другой движется со скоростью 8/10 скорости света. Ни один из них не должен быть *особенным* и его часы не должны замедляться относительно другого.

Чтобы разрешить этот парадокс, нужно рассмотреть некоторые дополнительные странные последствия того факта, что скорость света в вакууме *никогда не меняется*.

# Сокращение длины


Для последующего мысленного эксперимента направим наши часы-интерферометр в том же направлении, в котором движется космический корабль. Теперь сконцентрируемся на проходимом светом расстоянии:

$$T_{clock} = \frac {2d} c \rightarrow\\
d = \frac {T_{clock}c} 2$$


## Наземный наблюдатель

Предположим, что вы наблюдаете за космическим кораблем с земли и не знаете, каков линейный размер установленых там часов. Вы смотрите на пролетащий мимо космический корабль и фиксируете длину часов как $d_0$. Часы ориентированы так, чтобы излучатель импульсов был обращен вперед, и когда источник испускает импульс, вы видите, как импульс мчится впереди излучателя. По мере того как космический корабль движется быстрее, становится ясно, что импульсу потребуется все больше и больше времени, чтобы добраться до зеркальной стенки, потому что она удаляется со скоростью движения корабля. Так как космический корабль никогда не движется быстрее света, импульс в конечном итоге достигнет отражателя. 

Скорость светового импульса относительно космического корабля составляет $c-v$. Поэтому время, необходимое для того, чтобы импульс достиг зеркала, определяется: 

$$T_{forward} = \frac {d_0} {c-v}$$

Затем, когда импульс отскакивает назад, относительные скорости суммируется, и время:

$$T_{backward} = \frac {d_0} {c+v}$$

И находим суммарное время:

$$T_{observed} = T_{forward} + T_{backward} = d_0 \left ( \frac 1 {c-v} + \frac 1 {c+v} \right ) = \frac {2 d_0 c} {c^2 - v^2}$$


## Наблюдаемая длина

Вы, наблюдая с земли, можете быстро рассчитать длину часов, основываясь на том, сколько времени, по вашему мнению, требуется для завершения одного тика:

$$d_0 = \frac {T_{observed}(c^2 - v^2)} {2c}$$

Но вы, проницательный наблюдатель, уже знаете, что на самом деле не можете наблюдать за временем на корабле. Вы помните, что время на космических кораблях течет по-разному и что вы должны быть в состоянии предсказать наблюдаемое время на основе скорости космического корабля и длины часов, измеренных, когда он был неподвижен. Итак, начнем с уравнения тактовой частоты:

$$\frac {f_{observed}} {f_{clock}} =  \sqrt {1-\frac {v^2}{c^2}}$$

Выразим через период:

$$\frac {T_{observed}} {T_{clock}} =  \frac {1} {\sqrt {1-\frac {v^2}{c^2}}} = \gamma$$

$\gamma$ называют Лоренц фактором. Подставляя известное соотношение между $T_{clock}$ и $d$:

$$\frac {T_{observed}} {\frac {2d} {c}} =  \gammaф$$

$$ T_{observed} = \frac {2d}{c }\gamma$$

Теперь мы можем переписать наш расчет наблюдаемой длины, строго исходя из известной исходной длины и скорости космического корабля!

$$d_0 = \frac {\frac {2d}{c \sqrt{1-\frac {v^2}{c^2}}}(c^2 - v^2)} {2c} = \frac {d \left ( 1-\frac {v^2}{c^2} \right )}{\sqrt{1-\frac {v^2}{c^2}}} = \frac d \gamma$$

### Что только что произошло?

Поскольку мы *точно знаем*, насколько время замедляется на космическом корабле, и мы *знаем*, сколько времени требуется, на один шаг установленных там часов, то выходит, что часы должны **сжиматься** вдоль оси, параллельной движению космического корабля, чтобы все работало должным образом. Разумеется, это касается не только часов - весь космический корабль и все в нем сжимается точно так же. 

Правильно: вы *наблюдаете*, что космический корабль короче, чем должен быть в $\sqrt{1-\frac {v^2}{c^2}}$ раз!

Итак, теперь мы знаем две странные вещи о быстрых космических кораблях:

$$\frac {T_{stationary}} {T_{spaceship}} = \frac {T_{observed}} {T_{clock}} = \gamma$$

и 

$$\frac {d_{stationary}} {d_{spaceship}} = \frac d \gamma$$

Напомним: $T_{stationary}$ относится к продолжительности временного интервала, который происходит на космическом корабле, наблюдаемом с земли, в то время как $T_{spaceship}$ - это продолжительность интервала, измеренного на борту космического корабля. Кроме того, $d_{stationary}$ относится к длине движущегося космического корабля (или его части), измеренной вдоль направления его движения стационарным наблюдателем, в то время как $d_{spaceship}$ - это длина того же космического корабля (или его части), измеренная кем-то, кто путешествует вместе с ним.

Небольшой пример: помните космический корабль, который движется со скоростью 8/10 скорости света? Давайте посмотрим, насколько он уменьшится:

```python
from math import *
c = 3.00E8 # скорость света, метров в секунду
v = 0.8*c  # очень быстрый космический корабль
dr = sqrt(1 - v**2/c**2)
print("Лоренц фактор длины корабля: ", dr)
```

Это соотношение длин означает, что если бы космический корабль был, скажем, длиной 100 метров и двигался со скоростью 8/10 скорости света, наземный наблюдатель уверенно измерил бы его длину как 60 метров.

# Одновременность

Поскольку мы рассмотрели странные обстоятельства, при которых длины должны изменяться по мере того, как космический корабль движется быстрее, аналогичный метод может быть использован для выяснения того, происходят ли события на космическом корабле, одновременно с неподвижным наблюдателем. Это вопрос **одновременности** и будет измеряться с помощью нового прибора: **одновременометра**.

## Измеритель одновременности

Представим себе новую версию нашего интерферометра, в которой излучатель света находится в середине вакуумной камеры. Он излучает световой импульс в обоих направлениях (вперед и назад) одновременно. Вместо зеркал на каждой стенке стоит по детектору. Очевидно, что когда излучатель производит пару импульсов, они оба будут двигаться со скоростью света, $c$, и прибудут на свои соответствующие концы в одно и то же время, и сигнал от обоих детекторов вернется в центр измерителя одновременности в одно и то же время. 

## Наземный наблюдатель

Предположим, вы находитесь на земле, наблюдая за установленным на борту быстрого космического корабля одновременометром. Его длина с земли кажется равной $d_0$, а расстояние от излучателя до переднего конца, очевидно, составляет $d_0/2$.

Скорость импульса направленного вдоль движения кора бля, представляет собой разницу между $c$, скоростью самого светового импульса, и $v$, скоростью космического корабля. Поэтому время, необходимое для того, чтобы импульс достиг передней стенки, равно:

$$t_f = \frac {d_0/2} {c-v}$$

Скорость импульса направленного против движения корабля получается суммированием $c$ и $v$. Поэтому время, необходимое для достижения датчика на задней стенке, равно:

$$t_b = \frac {d_0/2} {c+v}$$

Поскольку эти два времени явно различны, два импульса поступают на соответствующие концы одновременометра *в разное время* с точки зрения стационарного наблюдателя. На самом деле разница во времени между двумя событиями, которые происходят одновременно в движущейся системе отсчета, составляет:

$$\Delta t_0 = t_f - t_b = \frac {d_0/2} {c-v} - \frac {d_0/2} {c+v} = \frac {d_0 v} {c^2 - v^2} = \frac {d_0 v} {c^2 \left (1-\frac {v^2}{c^2} \right )}$$

Поскольку $d_0$ определяется как оценка длины одновременометра стационарным наблюдателем, мы можем использовать соотношение сокращения длины, чтобы выразить его через $d$, которую использует космонавт:

$$d_0 = d \sqrt {1 - \frac {v^2}{c^2}}$$

И после подстановки:

$$\Delta t_0 = \frac {d v \sqrt {1 - \frac {v^2}{c^2}}} {c^2 \left (1-\frac {v^2}{c^2} \right )} = \frac {d \left(\frac {v} {c^2}\right)} {\sqrt{1-\frac {v^2}{c^2}}} $$

### Что это значит

Если система отсчета (СО) движется, то события, происходящие в ее передней части, задерживаются на $\Delta t_0$. Степень задержки зависит от того, насколько далеко впереди происходит событие, и от скорости движущегося кадра.

Мы также знаем, что замедление времени зависит исключительно от скорости СО:

$$\Delta t_0 = \Delta t\,  \gamma$$

Наблюдаемая временная задержка события является функцией временной задержки движущейся СО до этого события, положения события (в движущейся СО) и скорости движущейся СО:

$$\Delta t_0 = \frac {\Delta t + d \left ( \frac v {c^2} \right )} {\sqrt{1-\frac {v^2}{c^2}}}$$

Аналогично мы должны быть в состоянии написать выражение для наблюдаемой позиции события в движущейся СО. Начиная с нашего уравнения сокращения длины и простой формулировки положения в зависимости от наблюдаемой скорости и времени:

$$d_0 = d \sqrt{1 - \frac {v^2}{c^2}} + v \Delta t_0$$

преобразуем параметр $\Delta t_0$, используя новое уравнение замедления времени, приведенное выше:

$$d_0 = 
d \sqrt{1 - \frac {v^2}{c^2}} + \frac {v \left( \Delta t + d \left ( \frac v {c^2} \right ) \right )} {\sqrt{1- \frac {v^2}{c^2}}} = 
\frac {d - d \frac {v^2}{c^2} + v \Delta t + d  \frac {v^2}{c^2}} {\sqrt{1- \frac {v^2}{c^2}}} =
\frac {d  + v \Delta t } {\sqrt{1- \frac {v^2}{c^2}}}$$

## Преобразования Лоренца

Мы пришли к условному выражению преобразований Лоренца, которые могут быть использованы для преобразования события в пространстве-времени из движущейся СО ($x^{\prime}$, $t^{\prime}$) в соответствующее время $t$ для стационарного наблюдателя, 

$$t = \frac {t^{\prime} + x^{\prime} \left ( \frac v {c^2} \right )} {\sqrt{1-\frac {v^2}{c^2}}}$$

и в соответствующее положение $x$:

$$x = \frac {x^{\prime} + v t^{\prime}} {\sqrt{1-\frac {v^2}{c^2}}}$$

Оба этих выражения предполагают, что движение происходит в направлении $x$, тогда координаты $y$ и $z$ не требуют никакого преобразования.

# Парадокс близнецов



Парадокс близнецов предполагает сценарий, в котором космонавт отправляется в путешествие со скоростью близкой к скорости света, оставляя своего двойника на Земле. Из-за эффекта замедления времени, возникающего при путешествии на высоких скоростях, космонавт возвращается моложе своего близнеца. 

Для того, чтобы окончательно сформировать парадокс вам нужно только спросить, что делает космонавта таким особенным? Почему близнеца на Земле нельзя рассматривать как движущегося, и почему роли не меняются местами, делая космического путешественника тем более старым в конце путешествия? Симметрия ситуации, по-видимому, требует, чтобы ни один из близнецов не был старше другого, когда путешествие закончится!

Разрешение этого очевидного парадокса зависит от правильного использования преобразований Лоренца, а не от упрощенного применения эффекта замедления времени.

## Преобразования Лоренца

Преобразования Лоренца могут быть использованы для преобразования пространственно-временных координат события в движущейся СО ($x^{\prime}$ и $t^{\prime}$) в соответствующие пространственно-временные координаты в неподвижном СО наблюдателя ($x$ и $t$):

$$t = \frac {t^{\prime} + x^{\prime} \left ( \frac v {c^2} \right )} {\sqrt{1-\frac {v^2}{c^2}}}$$

$$x = \frac {x^{\prime} + v t^{\prime}} {\sqrt{1-\frac {v^2}{c^2}}}$$



## Сценарий парадокса Близнецов

Давайте создадим реальный сценарий с нашими бесстрашными близнецами, чтобы точно увидеть, что произойдет.

Начнем с некоторых предположений:
* Для простоты предположим, что космический корабль способен мгновенно ускоряться до 8/10 скорости света (ой!).
* Пункт назначения - звезда, находящаяся на расстоянии двух световых лет.
* На Земле и в космическом корабле есть часы.
* В пункте назначения есть часы, которые синхронизированы с земными часами. Это означает, что если, например, земные часы отправят радиосигнал в полночь 1 января 2050 года, часы звезды назначения получат этот сигнал Земли в полночь 1 января 2052 года.
* Близнец на Земли, наблюдатель на удаленной звезде и космонавт запускают свои секундомеры в момент вылета (1 января 2050 года).

## Что видит близнец на Земле

В полночь 1/1/2050 космонавт покидает Землю со скоростью 8/10 скорости света. Хотя земной близнец не может непосредственно наблюдать за тем, что происходит на далекой звезде, он знает, что космический корабль предназначен для торможения до остановки там и немедленно вернется на Землю со скоростью 8/10 скорости света. Таким образом, это простая проблема времени и расстояния, и время возврата легко предсказать:

```python
dly = 2       # удаленная звезда находится на расстоянии 2 световых лет
c = 3.00E8    # скорость света, метров в секунду
year = 365.25 * 24 * 60 * 60  # секунд в году
v = 8/10 * c
d = dly * c * year  # расстояние до удаленной звезды в метрах

T = 2*d/v     # время пуешествия в секундах для поездки туда и обратно со скоростью 8/10 *C
print("Время прибытия (в перспективе Земли): {0:.4} секунды.".format(T))
```

```python
# а сколько в годах?
```

## Что видит наблюдатель с удаленной звезды

В полночь, 1/1/2050, удаленный наблюдатель запускает свой секундомер. Зная, что космонавт должен был покинуть Землю в это время, наблюдатель ожидает, что космонавт прибудет несколько позже, чем через два года:

```python
Tstar = d/v 
print("Время прибытия (перспектива удаленной звезды): {0:.4} секунды.".format(Tstar))
```

```python
# а сколько в годах?
```

Пока все кажется тривиальным и очевидным.

## Что видит космонавт

Но теперь ситуация начинает выглядеть немного странно. Сразу после запуска секундомера в полночь 1/1/2050 астронавт наблюдает, как Земля удаляется со скоростью 8/10 скорости света. Точно так же удаленная звезда приближается с той же скоростью. Из-за этого относительного движения астронавт наблюдает сокращение расстояния между Землей и далекой звездой! Кроме того, космонавт уверен, что на удаленной звезде будут наблюдать время прибытия $T_{star}=d/v$, поэтому он может предсказать, что скажет их секундомер, когда он туда доберется:

```python
from math import *

gamma = lambda v,c: 1/sqrt(1 - v**2/c**2)
t_lorentz = lambda xp,tp,v,c: (tp + xp*(v/c**2))*gamma(v,c)

Tastronaut_star = t_lorentz(d, Tstar, -v, c)  # -v потому что видимое движение звезды противоположно движению космонавта
print("Время прибытия на удаленную звезду с точки зрения космонавта: {0:.4} seconds.".format(Tastronaut_star))
```

```python
# а сколько в годах?
```

Космонавта предупреждали, что путешествие займет не так много времени, как ожидалось с точки зрения Земли, и пока это кажется правдой.

## Космонавт возвращается на Землю

На обратном пути ситуация ничем не отличается: пропасть между звездой и Землей, кажется, снова сжимается и движется в направлении, противоположном движению космического корабля. Наблюдатель с Земли увидит, что обратный путь займет $T_{star}=d/v$ секунд, поэтому легко предсказать время прибытия:

```python
from datetime import timedelta, date
Tastronaut_earth = Tastronaut_star + t_lorentz(d, Tstar, -v, c)
print("Время прибытия на Землю (с точки зрения космонавта): {0:.4} секунды.".format(Tastronaut_earth))
# перевести в даты
dta = timedelta(seconds=int(Tastronaut_earth))  # время в пути с точки зрения космонавта
dte = timedelta(seconds=int(T))                 # время в пути с точки зрения Земли
Dstart = date(2050,1,1)                         # дата начала миссии
print("Дата прибытия (космонавт): {}".format(Dstart + dta))
print("Дата прибытия (близнец на Земле): {}".format(Dstart + dte))
```

Таким образом, предсказанное замедление времени все еще происходит для путешествующего астронавта. Почему? Если бы астронавт мог наблюдать за часами на Земле и на звезде, не показалось бы, что они также идут медленнее?

Ответ, конечно, да, но относительность одновременности бросает тень на проблему. Как только космонавт ускорится до скорости своего полета, он засвидетельствует, что секундомеры на Земле и на удаленной звезде *больше не синхронизированы*. На самом деле, будет казаться, что секундомер на удаленной звезде был запущен *раньше*, чем секундомер на Земле. Таким образом, когда космонавт прибудет к удаленной звезде, секундомер сообщит о более длительном времени путешествия. На обратном пути ситуация меняется на противоположную, и секундомер Земли будет казаться запущенным раньше секундомера звезды.

Если этот результат кажется бессмысленным, вернитесь назад и подумайте об эффекте одновременности. Как только вы поймете, как два события, которые явно одновременны в одной системе отсчета, могут быть явно **не** одновременными для наблюдателя в другой системе отсчета, тогда утверждение о секундомерах будет иметь смысл!
